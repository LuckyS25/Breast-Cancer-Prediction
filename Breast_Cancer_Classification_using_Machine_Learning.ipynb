Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection & Processing

```python
# loading data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
# loading the data to a panda frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print first five row of dataframe
data_frame.head()
```

```python
# addingg the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# pprint last 5 rows of data frame
data_frame.tail()
```

```python
# number of rows and thr column in thr dataset
data_frame.shape
```

```python
# getting some information about th data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measure about the data
data_frame.describe()
```

```python
# checking the distribution of the target variale
data_frame['label'].value_counts()
```

1 represent Benign
0 represent Malignant

```python
data_frame.groupby('label').mean()
```

Seperating the feature and target

```python
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logostic regression model using training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data =',training_data_accuracy)
```

```python
# accuracy on test data
X_train_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_train_prediction)
```

```python
print('Accuracy on test data',test_data_accuracy)
```
